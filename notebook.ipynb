### Project Overview:
This Tableau-driven BI project aims to analyze a real-world supply chain dataset for Just In Time. Through interactive Tableau dashboards and an executive report, the project seeks to identify supply chain challenges, uncover inefficiencies, and propose strategic business improvements.

## Issues and Concerns:

### The Problem:
After running initial exploratory data analysis (EDA) on a sample of the data provided, it is clear that the trend of Warehouse Inventory was dramatically dropped after September, 2017.
![image](image.png)


### Problem Exploration
On further exploration, it was found that In 2017, sales dropped notably from September to December. For example, Cameras dominated October largest proportion of Sales sales at 60.3% \$58K but declined to 14.3%, ($12.2K) in November. The sudden and significant decline in gross sales for all products is unusual, raising concerns about possible organizational responsibility or structural issues. The decline in warehouse inventory has significantly impacted gross sales.

### Recommendation:
Immediate consideration is required for further investigation as the USA stands out as the largest country by the quantity of orders. To address the situation, conduct a comprehensive analysis of the organizational responsibilities, sales processes, marketing strategies, inventory management, and customer feedback.

## Key Insights:
- The Latin America and Europe have the highest Gross Sales roughly \$1.8M in both Markets and Profit of ~\$1.14M in Europe in ~$1.18M in Latin America.
- Perfect Fitness Perfect Rip Desk was the top product by profit of ~$631K. The product was sold in USA with 12,952 orders.
- During Analysis, Countries Like Germany, France, Mexico and USA were identified as top customer countries, among which USA is the largest market of around 1,365 customers.
- Average Inventory cost was roughly same among the top products.
- Africa is the lowest market of Sales with \$0.41M and profit of $0.7M

## Next Steps:
1. The stakeholders must look into the matter why Warehouse Inventory dropped drastically resulting in declined monthly Sales. A thorough investigation is needed to pinpoint the root cause of the Warehouse inventory decline and adapt strategies to restore sales momentum and competitiveness.
2. The African market has shown the lowest sales. To explore potential growth, further analysis is needed to identify opportunities and countries with significant buying power.
